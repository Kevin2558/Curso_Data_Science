```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
X,y = load_breast_cancer(return_X_y=True,as_frame=True)
X = pd.DataFrame(X)
y = pd.DataFrame(y)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0)
```

```python
X
```

```python
X.dtypes
```

```python
from sklearn.preprocessing import StandardScaler, RobustScaler
from sklearn.pipeline import Pipeline
```

```python
pre = Pipeline([
    ('scaler', RobustScaler())
])
```

```python
pipe_nb = Pipeline([
    ('pre', pre),
    ('nb', GaussianNB())
])
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
pipe_nb.fit(X_train, y_train)
y_pred = pipe_nb.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
```

```python
pipe_gb = Pipeline([
    ('pre', pre),
    ('gb', GradientBoostingClassifier(
        n_estimators=100,
        learning_rate=1.0,
        max_depth=3,
        random_state=42
    ))
])
```

```python
pipe_gb.fit(X_train, y_train)
y_pred = pipe_gb.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
import kagglehub
import pandas as pd

# Download latest version
path = kagglehub.dataset_download("yasserh/titanic-dataset")

print("Path to dataset files:", path)
```

```python
data = pd.read_csv(path + '/Titanic-Dataset.csv')
```

```python
data['Age'] = data['Age'].fillna(data['Age'].mean())
data['Embarked']= data['Embarked'].fillna(data['Embarked'].mode()[0])
```

```python
data
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
data['Sex'] = LabelEncoder().fit_transform(data['Sex'])
data['Embarked'] = LabelEncoder().fit_transform(data['Embarked'])
```

```python
data
```

```python
y = data['Survived']
X = data.drop(columns=['Survived','Name','Ticket','Cabin','PassengerId'])
```

```python
num = ['Age','SibSp','Parch','Fare']
cat = ['Sex','Embarked','Pclass']
```

```python
X = X[num+cat]
```

```python
from sklearn.compose import ColumnTransformer
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
prep = ColumnTransformer([
    ('num', RobustScaler(), num),
    ('cat', OneHotEncoder(drop='first'), cat)
])
```

```python
prep = ColumnTransformer([
    ('num', RobustScaler(), num),
    ('cat', 'passthrough', cat)
])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=0,stratify=y)
```

```python
pipe_xgb_gpu_2 = Pipeline(
    [
        ('prep', prep),
        ('xgb', XGBClassifier(
            n_estimators=500,
            learning_rate=0.05,
            max_depth=3,
            enable_categorical=True,
            random_state=42,
            sub_sample = 0.5,
            objective = 'binary:logistic',
            device = 'gpu',
            colsample_bytree = 0.5,
            eval_metric = 'logloss'
        ))
    ]
)
```

```python
pipe_xgb_gpu_2.fit(X_train, y_train)
y_pred = pipe_xgb_gpu_2.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
pipe_lr = Pipeline(
    [
        ('prep', prep),
        ('lr', LogisticRegression(
            max_iter=1000,
            random_state=42
        ))
    ]
)
```

```python
pipe_lr.fit(X_train, y_train)
y_pred = pipe_lr.predict(X_test)
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```
